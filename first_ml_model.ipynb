```python
# Step 1: Import the basic libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Step 2: Create data (hours studied vs exam score)
hours = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)   # input (X)
scores = np.array([30, 40, 50, 60, 70, 80])           # output (y)

# Step 3: Create the model and train it
model = LinearRegression()
model.fit(hours, scores)

# Step 4: Predict a new student's score
new_hours = np.array([[7]])        # student studies 7 hours
predicted_score = model.predict(new_hours)

# Step 5: Show result
print(f"If a student studies for 7 hours, predicted score = {predicted_score[0]:.2f}")

# Step 6: Visualize the learning
plt.scatter(hours, scores, color='blue', label='Actual')
plt.plot(hours, model.predict(hours), color='red', label='Learned Line')
plt.xlabel('Hours Studied')
plt.ylabel('Score')
plt.legend()
plt.show()
```
