```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# data
data = {
    'Size': [2100,1600,2400,1416,3000],
    'Bedrooms': [3,2,4,2,4],
    'Age': [20,15,30,20,8],
    'Price': [400,330,369,232,540]
}
df = pd.DataFrame(data)

# train
X = df[['Size','Bedrooms','Age']]
y = df['Price']
model = LinearRegression()
model.fit(X, y)

# predict
new_house = pd.DataFrame({'Size':[2500], 'Bedrooms':[3], 'Age':[10]})
print(f"Predicted price: ${model.predict(new_house)[0]*1000:.2f}")
```
