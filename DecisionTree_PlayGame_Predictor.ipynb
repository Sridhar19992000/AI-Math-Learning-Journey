```python
# Step 1: Import Libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
# Step 2: Create a small dataset
data = {
    'Weather': ['Sunny', 'Sunny', 'Overcast', 'Rainy', 'Rainy', 'Rainy', 'Overcast', 'Sunny', 'Sunny', 'Rainy', 'Sunny', 'Overcast', 'Overcast', 'Rainy'],
    'Temperature': ['Hot', 'Hot', 'Hot', 'Mild', 'Cool', 'Cool', 'Cool', 'Mild', 'Cool', 'Mild', 'Mild', 'Mild', 'Hot', 'Mild'],
    'Play': ['No', 'No', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 'Yes', 'Yes', 'Yes', 'Yes', 'No']
}

df = pd.DataFrame(data)
# print(df)

from sklearn.preprocessing import LabelEncoder


# Step 3: Convert text data to numbers
le_weather = LabelEncoder()
le_temp = LabelEncoder()
le_play = LabelEncoder()

df['Weather_n'] = le_weather.fit_transform(df['Weather'])
df['Temperature_n'] = le_temp.fit_transform(df['Temperature'])
df['Play_n'] = le_play.fit_transform(df['Play'])

# print(df)

# Step 4: Train the Decision Tree model
X = df[['Weather_n', 'Temperature_n']]   # Features (inputs)
y = df['Play_n']                         # Target (output)

model = DecisionTreeClassifier()
model.fit(X, y)
from sklearn import tree
plt.figure(figsize=(10,6))
tree.plot_tree(model, feature_names=['Weather', 'Temperature'], class_names=['No', 'Yes'], filled=True)
plt.show()

# Step 6: Predict for new data
# Let's say: Sunny (2), Mild (2)
prediction = model.predict([[2, 2]])
result = le_play.inverse_transform(prediction)
print("Prediction for Sunny & Mild:", result[0])


```
